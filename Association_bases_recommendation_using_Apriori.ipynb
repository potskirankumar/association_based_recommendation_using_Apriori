```python
import random
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
```

```python
customers = range(1, 101)
```

```python
items = [
    "Phone", "Case", "Screen Protector", "Laptop", "Mouse", "Keyboard",
    "Charger", "Earphones", "Docking Station", "Mousepad", "Tablet", "Power Bank",
    "Smart Watch", "Speakers", "Camera", "Tripod", "SD Card", "Hard Drive"
]
```

```python
data = []
for _ in range(100):
    customer = random.choice(customers)
    num_items = random.randint(2, 5)
    purchased_items = random.sample(items, num_items)
    data.append({'CustomerID': customer, 'Items': purchased_items})
```

```python
transactions_df = pd.DataFrame(data)
```

```python
transactions = []
for index, row in transactions_df.iterrows():
    for item in row['Items']:
        transactions.append({'CustomerID': row['CustomerID'], 'Item': item})
```

```python
transaction_df = pd.DataFrame(transactions)
```

```python
basket = transaction_df.pivot_table(index='CustomerID',
                                    columns='Item',
                                    aggfunc=lambda x: 1,
                                    fill_value=0)
```

```python
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
def recommend_items(user_id, user_transactions):
    user_cart = set(user_transactions)
    recommendations = rules[rules['antecedents'] == frozenset(user_cart)]
    recommended_items = recommendations['consequents'].apply(lambda x: list(x)).tolist()
    return recommended_items
```

```python
user_id = 81
user_transactions = transactions_df[transactions_df['CustomerID'] == user_id]['Items'].values[0]
```

```python
recommended_items = recommend_items(user_id, user_transactions)
```

```python
print(f"User {user_id}'s Cart: {user_transactions}")
print(f"Recommended Items for User {user_id}: {recommended_items}")
```
